# MarketSim Example

Below, I show how to simulate 10 minutes of E-mini S&P 500 transactions.

```python
%matplotlib inline
import matplotlib
from limit_order_book import MarketSim

# target volume is 600/23.3 per second
# assume approximately 50% of the trades cross
arrival_lambda = 1./((600/23.3)*(1./.50))

params = {'order_arrival_lambda': arrival_lambda,
          'price_sigma': .005,
          'tick_size': .0025,
          'quantity_mu': 1,
          'quantity_sigma': 1.1
          }
sim = MarketSim(**params)
sim.simulate_periods(600, verbose=False)
sim.show_stats()
```

```python
sim.plot_prices(time_scale=1,plot_params={'figsize':(16,4)})
```

```python
sim.get_transactions().set_index("time")['quantity'].plot(figsize=(16,4));
```
